```python
a=np.array([5,15,25,27,16,4,36,33,7])
plt.hist(a,bins=5) 
#bins shows the density 
#Using wider bins, where the densitys low, reduces nois due sampling randomness.using narrower bins, where the signal drowns the noise, gives greater precision to the density estimation. Thus varying the bin width within a histogram can be beneficial.
```

```python
data=[np.random.normal(0,std,100) for std in range(1,4)]
#rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True);
#patch_artist=color on or off
#vert=vertical or horizontal
#finding percentile
```
